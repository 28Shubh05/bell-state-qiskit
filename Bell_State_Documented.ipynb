
# üß© Bell State Demonstration

This notebook demonstrates how to create and measure a **Bell state** using Qiskit.
A Bell state is a maximally entangled two-qubit state and is fundamental in quantum computing and quantum information.

---


### Imports -

```python
# üìö Import necessary libraries
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
# üìö Import necessary libraries
from qiskit_aer import AerSimulator
# üìö Import necessary libraries
from qiskit.visualization import plot_histogram
```

```python
qr = QuantumRegister(2 , 'q')
cr = ClassicalRegister(2, 'c')
# ‚öôÔ∏è Create a quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(qr, cr)

# üñºÔ∏è Draw the quantum circuit
qc.draw("mpl")
```

### adding gates

```python
# Apply Hadamard gate to qubit 0 to create superposition
qc.h(0)
# Apply CNOT gate to entangle qubit 0 and qubit 1
qc.cx(0, 1)

qc.barrier()

# üñºÔ∏è Draw the quantum circuit
qc.draw("mpl")
```

```python
# Measure both qubits and store results in classical bits
qc.measure(qr, cr)

# üñºÔ∏è Draw the quantum circuit
qc.draw("mpl")
```

### run on simulator

```python
job = AerSimulator().run(qc , shots = 1024)
# Get the simulation result
job_result = job.result()
# Get the simulation result
counts = job_result.get_counts(qc)
print(counts)
```

```python
# üìä Visualize the measurement results as a histogram
plot_histogram(counts)
```
