# Single Qubit Implementation

## imports

```python
# üìö Import required libraries
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

## circiut building

```python
# ‚öôÔ∏è Create a quantum circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)
# Apply a Hadamard gate to create a superposition
qc.h(0)

# Measure the qubit into the classical bit
qc.measure(0, 0)

# üñºÔ∏è Draw the circuit diagram
qc.draw("mpl")
```

## running on simulator

```python
job = AerSimulator()
backend = job.run(qc)
# Get the simulation result
job_result = backend.result()
# Get the simulation result
counts = job_result.get_counts(qc)
# üìä Plot a histogram of results
plot_histogram(counts)
```
